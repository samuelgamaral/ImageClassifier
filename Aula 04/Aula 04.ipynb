```python
import cv2
import os

caminho_amigos = "/content/drive/MyDrive/Educação/Tecnologia/Ciência de Dados UECE/Módulo Intermediário – Algoritmos de Ciência de Dados (100 horas)/Aula 04/Fotos/Amigos"
caminho_familiares = "/content/drive/MyDrive/Educação/Tecnologia/Ciência de Dados UECE/Módulo Intermediário – Algoritmos de Ciência de Dados (100 horas)/Aula 04/Fotos/Familiares"

for pasta in [caminho_amigos, caminho_familiares]:
    for arquivo in os.listdir(pasta):
        img = cv2.imread(os.path.join(pasta, arquivo))
        img_redimensionada = cv2.resize(img, (100, 100))
        cv2.imwrite(os.path.join(pasta, arquivo), img_redimensionada)
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense

datagen = ImageDataGenerator(rescale=1./255, validation_split=0.3)
treino = datagen.flow_from_directory('/content/drive/MyDrive/Educação/Tecnologia/Ciência de Dados UECE/Módulo Intermediário – Algoritmos de Ciência de Dados (100 horas)/Aula 04/Fotos', target_size=(100, 100), batch_size=32, class_mode='binary', subset='training')
teste = datagen.flow_from_directory('/content/drive/MyDrive/Educação/Tecnologia/Ciência de Dados UECE/Módulo Intermediário – Algoritmos de Ciência de Dados (100 horas)/Aula 04/Fotos', target_size=(100, 100), batch_size=32, class_mode='binary', subset='validation')

modelo = Sequential([
    Conv2D(32, (3, 3), activation='relu', input_shape=(100, 100, 3)),
    MaxPooling2D((2, 2)),
    Flatten(),
    Dense(64, activation='relu'),
    Dense(1, activation='sigmoid')
])

modelo.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

modelo.fit(treino, validation_data=teste, epochs=30)
```

```python
import numpy as np
from tensorflow.keras.preprocessing import image

nova_foto = image.load_img('/content/drive/MyDrive/Educação/Tecnologia/Ciência de Dados UECE/Módulo Intermediário – Algoritmos de Ciência de Dados (100 horas)/Aula 04/Familiares Test/Gus_Van_Sant - Test/Gus_Van_Sant_0001.jpg', target_size=(100, 100))
foto_array = image.img_to_array(nova_foto) / 255.0
foto_array = np.expand_dims(foto_array, axis=0)

resultado = modelo.predict(foto_array)
if resultado[0][0] > 0.5:
    print("Familiar")
else:
    print("Amigo")
```
